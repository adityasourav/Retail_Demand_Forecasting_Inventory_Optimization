```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
comparison = pd.DataFrame({
    "Model": [
        "Random Forest",
        "ARIMA",
        "XGBoost"
    ],
    "MAE": [
        12.82,
        3169.64,
        1020.99
    ],
    "RMSE": [
        17.07,
        3644.91,
        1431.94
    ],
    "R2 Score": [
        0.868,
        -3.006,
        0.334
    ]
})

comparison
```

```python
plt.figure(figsize=(8,5))

plt.bar(comparison["Model"], comparison["R2 Score"])

plt.title("Model Comparison (R² Score)")
plt.ylabel("R² Score")
plt.xlabel("Models")

plt.show()
```

```python
best_model = comparison.loc[
    comparison["R2 Score"].idxmax()
]

print(best_model)
```

```python
print("""
Conclusion:

Random Forest achieved the best performance among all models.

R² Score : 0.868
MAE      : 12.82
RMSE     : 17.07

Therefore, Random Forest is selected as the final model for deployment.
""")
```
